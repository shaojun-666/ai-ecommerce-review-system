# 数据探索与分析
## AI电商评论分析系统 — 数据概览
本 notebook 用于探索原始评论数据集的基本统计信息和分布特征。

```python
import sys
sys.path.insert(0, '..')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['font.sans-serif'] = ['SimHei', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False

%matplotlib inline
```

```python
# 加载数据
df = pd.read_csv('../data/raw/reviews.csv') if False else pd.DataFrame()
print('数据集行数:', len(df))
print('数据集列数:', len(df.columns))
df.head()
```

```python
# 数据分布
if len(df) > 0:
    fig, axes = plt.subplots(1, 3, figsize=(15, 4))
    df['rating'].value_counts().sort_index().plot(kind='bar', ax=axes[0], title='评分分布')
    df['sentiment'].value_counts().plot(kind='pie', ax=axes[1], title='情感分布', autopct='%1.1f%%')
    df['platform'].value_counts().plot(kind='bar', ax=axes[2], title='平台分布')
    plt.tight_layout()
```

```python
# 评论文本长度分布
if len(df) > 0:
    df['text_length'] = df['content'].astype(str).str.len()
    plt.figure(figsize=(10, 4))
    plt.subplot(1, 2, 1)
    df['text_length'].hist(bins=50)
    plt.title('文本长度分布')
    plt.xlabel('长度')
    
    plt.subplot(1, 2, 2)
    df.boxplot(column='text_length', by='sentiment')
    plt.title('各情感类别文本长度')
    plt.suptitle('')
    plt.tight_layout()
```
